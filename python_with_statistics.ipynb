# Descriptive statistics

```python
import statistics
import array as arr

friendUsage = [2,32,43,21,65,76,12,87,43,31,56,2,97,43,23,81,53,17,65,35,43,]
sampleData = arr.array("i", friendUsage)
dataset = sampleData[0:5]
sampleDataList = dataset.tolist()

mean = statistics.mean(friendUsage)
mode = statistics.mode(friendUsage)
median = statistics.median(friendUsage)
pstdev = round(statistics.pstdev(friendUsage),2)
pvariance = round(statistics.pvariance(friendUsage),2)
srange = (max(friendUsage)-min(friendUsage))

variance = round(statistics.variance(sampleDataList),2)
stdev = round(statistics.stdev(sampleDataList),2)

print(mean)
print(mode)
print(median)
print(pstdev)
print(pvariance)
print(srange)
print(variance)
print(stdev)
```

# Displaying statistics data using matplotlib

# coding exercise 4.8

```python
import pandas as pd
import matplotlib.pyplot as plt

dFrame = pd.read_csv("population_by_country_2020.csv")
eastAsia = dFrame[dFrame["Country"].isin(["China", "South Korea", "North Korea", "Macau", "Hong Kong", "Tibet",
                                             "Taiwan", "Japan", "Mongolia"])]

eastAsia
```

```python
country = eastAsia["Country"]
population = eastAsia["Population (2020)"]

figGraph = plt.figure(figsize=(16,10))
plt.barh(country, population)
plt.title("2020 Population by Country (East Asia)", loc="center")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,10))
ax.barh(country,population)
ax.set_title("2020 Population by Country (East Asia)", loc="center",)

# remove axes splines
for s in ["top", "bottom", "left", "right"]:
    ax.spines[s].set_visible(False)
    
# add to annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5, str(round(i.get_width(),2)), fontsize=10, fontweight="bold", color="grey")
    
# remove x, y ticks
ax.xaxis.set_ticks_position("none")
ax.yaxis.set_ticks_position("none")

# add padding/spaces between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

# add x, y gridlines
ax.grid(b=True, color="grey", linestyle="-.", linewidth=0.5, alpha=0.2)

#show top values
ax.invert_yaxis()

plt.show()
```
